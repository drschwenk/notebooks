```python
def read_annotations(image_name):
    filename = "/Users/schwenk/wrk/vision/datasets/shining3/annotations/%s.json" % image_name
    with open(filename) as f:
        return json.loads(f.read())
```

```python
def flip_xy(element):
    element_np = np.array(element)
    flipped = np.copy(element_np)
    flipped[:, 0] = element_np[:, 1]
    flipped[:, 1] = element_np[:, 0]

    return flipped.tolist()
```

```python
def next_relationship_id(annotations):
    total = 0
    try:
        total += length(annotations['relationship']['interObject']['linkage'])
    except:
        pass

    try:
        total += length(annotations['relationship']['intraObject']['linkage'])
    except:
        pass

    try:
        total += length(annotations['relationship']['intraObject']['label'])
    except:
        pass

    return 'L' + str(total)
```

```python
def validate_arrowheads(image_name, annotations):

    def check_for_orphan_heads(all_arrow_heads, arrows_assoc_heads):
        all_assoc_head = arrows_assoc_heads.values()
        all_assoc_head_flat = [val for sublist in all_assoc_head for val in sublist]
        print(all_arrow_heads, all_assoc_head)
        all_attached = set(all_arrow_heads) == set(all_assoc_head_flat)
        return all_attached

    def check_many_heads(arrows_assoc_heads):
        for arrow, head in arrows_assoc_heads.items():
            if len(head) > 2:
                return False
        return True

    all_arrow_heads = []
    arrows_assoc_heads = {}

    try:
        for arrow, desc in annotations['arrows'].items():
            arrows_assoc_heads[arrow] = desc['arrowHeads']
        for head, desc in annotations['arrowHeads'].items():
            all_arrow_heads.append(head)
    except KeyError:
        warnings.warn("there are no arrowheads in  %s" % image_name)

    if not check_for_orphan_heads(all_arrow_heads, arrows_assoc_heads):
        warnings.warn("there are orphaned arrowheads in %s" % image_name)

    if not check_many_heads(arrows_assoc_heads):
        warnings.warn("there are too many arrowheads in %s" % image_name)
```
